```python
#----------------------------------------------------------------------
# OTTIENE SIGLA PROVINCIA DA MAPQUEST
#----------------------------------------------------------------------
import requests
import pprint
from tqdm import tqdm_notebook as tqdm
import pandas as pd
import requests
import json

api_errors = []
df = pd.read_csv('ST_colonnine_IT.csv', sep=';')
df = df.reset_index()

prov = []
i = 0 
f = open('ST_colonnine_IT_Prov.csv', 'a')
f.write('idStation;speed;location;lat;lng;Sigla\n')

for index, adr in df.iterrows():

    address = adr['location']

    key = "wrHRhgF0SfY5PPIYfoF07HuPkyjGckhu"
    geocode_url = f"http://www.mapquestapi.com/geocoding/v1/address?key={key}&location={address}"
   
    try:
        response = requests.get(geocode_url)
        geo = json.loads(response.text)

        for r in geo['results']:    
            for l in r['locations']:
                if 'IT' in l['adminArea1']:
                    line =  str(adr.idStation) + ';' + str(adr.speed) + ';' + str(adr.location) \
                            + ';' + str(adr.lat) + ';' + str(adr.lng) + ';' + str(l['adminArea4']) + '\n'
                    f.write(line)
                    break

    except Exception as e:
        e_text = 'error : ', address, e
        print(e_text)
        api_errors.append(e_text)
        df_api_errors = pd.DataFrame(api_errors)
        df_api_errors.to_csv('error_log.csv', mode='a', index=False, header=False, sep=';')
        api_errors = []

f.close()
#     i += 1 
#     print(i)

# df['Provincia'] = prov
# df.to_csv('ST_colonnine_IT_Prov.csv', sep=';', index=False, header=True)

```

```python
#----------------------------------------------------------------------
# DA SIGLA A NOME PROVINCIA
#----------------------------------------------------------------------
dfl = pd.read_csv('ST_colonnine_IT_Prov.csv', sep=';')

names = {'Milan':'MI', 'Genoa':'GE', 'Metropolitan City of Florence':'FI', 'Monza and Brianza':'MB', \
         'Burggrafenamt - Burgraviato':'BZ', 'Pustertal - Val Pusteria':'BZ', 'Mantua':'MN' }

for n in names:
   dfl['Sigla'] = dfl['Sigla'].str.replace(n, names[n])

#dfl = df.reset_index()
dfd = pd.read_csv('Sigla_Provincia.csv', sep=';')

f = open('ST_colonnine_IT_Prov_Nome.csv', 'a')
f.write('idStation;speed;location;lat;lng;Sigla;Provincia\n')

b = open('ST_colonnine_IT_Prov_Nome_BAD.csv', 'a')
b.write('idStation;speed;location;lat;lng;Sigla\n')

for index, adr in dfl.iterrows():
    sigla = str(adr['Sigla'])
    prov = dfd.query('SIGLA in @sigla')
    if not prov['PROVINCIA'].empty:
        p = str(prov['PROVINCIA'].values[0]).lower()
        line =  str(adr.idStation) + ';' + str(adr.speed) + ';' + str(adr.location) \
        + ';' + str(adr.lat) + ';' + str(adr.lng) + ';' + str(sigla) + ';' + str(p) + '\n'
        f.write(line)
    else:
        line = str(adr.idStation) + ';' + str(adr.speed) + ';' + str(adr.location) \
        + ';' + str(adr.lat) + ';' + str(adr.lng) + ';' + str(sigla) + '\n'
        b.write(line)

f.close()
b.close()
```
